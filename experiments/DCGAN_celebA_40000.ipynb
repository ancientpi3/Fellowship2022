```python
drive.mount('/content/gdrive')
filename = "/content/gdrive/MyDrive/celeba_train.zip"

with ZipFile(filename, "r",) as zip:
  zip.extractall(path="/content/celeba_train")
```

```python
def trial(x_train):
  d_model = DCGAN.discriminator()
  g_model = DCGAN.generator()
  gan_model = DCGAN.GAN(d_model,g_model)
  gan_model.summary()
  DCGAN.train(d_model,g_model,gan_model, x_train, epochs=390, batches_per_epoch=64, verbose=1, print_every=20)
  return g_model
```

```python
folder_path_train="/content/celeba_train"
x_train = DCGAN.folder_to_numpy(folder_path_train,take=40000)
```

```python
plt.imshow(x_train[0])
```

```python
g_model = trial(x_train)
```

```python
g_model.save("celeba_generator_40000_v3.h5")
```
